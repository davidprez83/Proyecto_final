### **Ejemplos para pasar a UTC**

```python
from datetime import datetime, timedelta
from astropy.time import Time
import astropy.units as u

# ================================================================
# 1. DATETIME: Crear tiempos en UTC
# ================================================================
print("="*60)
print("1. DATETIME: Crear tiempos en UTC")
print("="*60)

# Crear un datetime UTC específico
t0_dt = datetime(2029, 4, 13, 0, 0, 0)  # UTC por defecto en datetime
print(f"Datetime t0: {t0_dt}")
print(f"Tipo: {type(t0_dt)}")

# Representación ISO format
print(f"ISO format: {t0_dt.isoformat()}")

# Componentes individuales
print(f"Año: {t0_dt.year}, Mes: {t0_dt.month}, Día: {t0_dt.day}")
print(f"Hora: {t0_dt.hour}, Minuto: {t0_dt.minute}, Segundo: {t0_dt.second}")
```

```python
# ================================================================
# 2. OPERACIONES ARITMÉTICAS CON TIMEDELTA
# ================================================================
print("\n" + "="*60)
print("2. OPERACIONES ARITMÉTICAS CON TIMEDELTA")
print("="*60)

# Crear intervals de tiempo con timedelta
intervalo = timedelta(days=2, hours=3, minutes=30, seconds=45)
print(f"Intervalo: {intervalo}")
print(f"Total segundos: {intervalo.total_seconds()}")

# Suma de tiempos
t1_dt = t0_dt + intervalo
print(f"\nt0 + intervalo: {t1_dt}")

# Resta de tiempos
dt_diff = t1_dt - t0_dt
print(f"t1 - t0 = {dt_diff}")
print(f"Diferencia en segundos: {dt_diff.total_seconds()}")
print(f"Diferencia en horas: {dt_diff.total_seconds() / 3600}")
print(f"Diferencia en días: {dt_diff.days} días + {dt_diff.seconds} segundos")

# Multiples de timedelta
t2_dt = t0_dt + 5 * timedelta(hours=1)
print(f"\nt0 + 5 horas: {t2_dt}")
```

```python
# ================================================================
# 3. PARSING DE STRINGS A DATETIME
# ================================================================
print("\n" + "="*60)
print("3. PARSING DE STRINGS A DATETIME")
print("="*60)

# Diferentes formatos de strings
fecha_iso = "2029-04-13T12:30:45"
t_iso = datetime.fromisoformat(fecha_iso)
print(f"ISO format: {fecha_iso} -> {t_iso}")

# Parsing manual con strptime
fecha_str = "13/04/2029 14:30:00"
t_parsed = datetime.strptime(fecha_str, "%d/%m/%Y %H:%M:%S")
print(f"Custom format: {fecha_str} -> {t_parsed}")

# Otros ejemplos comunes
formatos = [
    ("2029-04-13", "%Y-%m-%d"),
    ("04-13-2029", "%m-%d-%Y"),
    ("13 April 2029", "%d %B %Y"),
    ("2029-04-13 23:59:59", "%Y-%m-%d %H:%M:%S"),
]

print("\nEjemplos de parsing con diferentes formatos:")
for fecha_str, fmt in formatos:
    t = datetime.strptime(fecha_str, fmt)
    print(f"  '{fecha_str}' -> {t}")
```

```python
# ================================================================
# 4. ASTROPY.TIME: Tiempo astronómico en UTC
# ================================================================
print("\n" + "="*60)
print("4. ASTROPY.TIME: Tiempo astronómico en UTC")
print("="*60)

# Crear Time en UTC
t0_astro = Time("2029-04-13 00:00:00", scale="utc")
print(f"Astropy Time t0: {t0_astro}")
print(f"Tipo: {type(t0_astro)}")

# Diferentes formatos de salida
print(f"\nFormatos de salida:")
print(f"  ISO: {t0_astro.iso}")
print(f"  ISOT: {t0_astro.isot}")
print(f"  yday: {t0_astro.yday}")
print(f"  jyear: {t0_astro.jyear}")

# Crear desde datetime
t0_from_dt = Time(t0_dt, scale="utc")
print(f"\nDesde datetime: {t0_from_dt.iso}")

# Crear desde diferentes formatos de string
t_formats = [
    Time("2029-04-13", scale="utc"),
    Time("2029-04-13T12:30:45", scale="utc"),
    Time("2029-04-13 14:30:00", scale="utc"),
]

print("\nDiferentes formatos de creación (strings ISO):")
for i, t in enumerate(t_formats):
    print(f"  Formato {i+1}: {t.iso}")
```

```python
# ================================================================
# 5. DÍAS JULIANOS (JULIAN DATE)
# ================================================================
print("\n" + "="*60)
print("5. DÍAS JULIANOS (JULIAN DATE)")
print("="*60)

# JD (Julian Date) - número de días desde mediodía del 1 de enero de 4713 BC
print(f"Julian Date (JD): {t0_astro.jd}")
print(f"Julian Date (JD) completo: {t0_astro.jd:.10f}")

# MJD (Modified Julian Date) - para números más pequeños
print(f"Modified Julian Date (MJD): {t0_astro.mjd}")
print(f"MJD completo: {t0_astro.mjd:.10f}")

# Diferencia
print(f"\nRelación: JD - 2400000.5 = MJD")
print(f"  {t0_astro.jd:.1f} - 2400000.5 = {t0_astro.mjd:.1f}")

# Crear a partir de JD
t_from_jd = Time(2462343.5, format="jd", scale="utc")
print(f"\nDesde JD=2462343.5: {t_from_jd.iso}")

# Crear a partir de MJD
t_from_mjd = Time(62343.0, format="mjd", scale="utc")
print(f"Desde MJD=62343.0: {t_from_mjd.iso}")

# Ejemplos de días julianos para diferentes fechas
print("\n" + "-"*60)
print("Ejemplos de JD para diferentes fechas:")
print("-"*60)
fechas_ejemplo = [
    "2000-01-01 12:00:00",
    "2020-01-01 00:00:00",
    "2025-05-11 00:00:00",
    "2029-04-13 12:30:45",
]

for fecha_str in fechas_ejemplo:
    t = Time(fecha_str, scale="utc")
    print(f"  {fecha_str:22} -> JD: {t.jd:15.5f}, MJD: {t.mjd:15.5f}")
```

```python
# ================================================================
# 6. OPERACIONES ARITMÉTICAS CON ASTROPY.TIME
# ================================================================
print("\n" + "="*60)
print("6. OPERACIONES ARITMÉTICAS CON ASTROPY.TIME")
print("="*60)

# Suma de unidades de tiempo
t_plus = t0_astro + 2 * u.day + 3 * u.hour + 30 * u.minute
print(f"t0 + 2 días + 3 horas + 30 min: {t_plus.iso}")

# Resta de tiempos
t1_astro = Time("2029-04-15 03:30:00", scale="utc")
diferencia = t1_astro - t0_astro
print(f"\nt1 - t0 = {diferencia}")
print(f"Diferencia en días: {diferencia.jd}")
print(f"Diferencia en segundos: {diferencia.sec}")

# Restar unidades específicas
t_minus = t1_astro - 1 * u.day
print(f"\nt1 - 1 día: {t_minus.iso}")

# Multiplicar por un escalar (crea un array)
t_array = t0_astro + [0, 1, 2, 3] * u.day
print(f"\nArray de tiempos (t0 + 0,1,2,3 días):")
for i, t in enumerate(t_array):
    print(f"  {i}: {t.iso}")

# Diferencia entre múltiples tiempos
tiempos = Time([
    "2029-04-13 00:00:00",
    "2029-04-14 12:00:00",
    "2029-04-15 06:30:00",
], scale="utc")

print(f"\nArray de tiempos:")
for t in tiempos:
    print(f"  {t.iso}")

# Diferencias entre el primero y los demás
print(f"\nDiferencias respecto al primero (en horas):")
for i, t in enumerate(tiempos):
    dt = (t - tiempos[0]).jd * 24  # convertir a horas
    print(f"  {i}: {dt:8.2f} horas")
```

```python
# ================================================================
# 7. EJEMPLO PRÁCTICO: Seguimiento orbital con tiempos
# ================================================================
print("\n" + "="*60)
print("7. EJEMPLO PRÁCTICO: Simulación orbital simplificada")
print("="*60)

# Parámetros de una órbita
epoch_str = "2029-04-13 00:00:00"
epoch = Time(epoch_str, scale="utc")
periodo_orbital = 90 * u.minute  # minutos en órbita

print(f"Época: {epoch.iso}")
print(f"Período orbital: {periodo_orbital.to(u.hour)}")

# Calcular posiciones en diferentes momentos
print(f"\nMomentos de máximo acercamiento (perigeo):")
print(f"{'Núm.':^6} {'ISO':^25} {'JD':^20} {'MJD':^20}")
print("-" * 72)

for n in range(6):
    # Tiempo del n-ésimo perigeo
    t_peri = epoch + n * periodo_orbital
    jd = t_peri.jd
    mjd = t_peri.mjd
    
    print(f"{n:^6} {t_peri.iso:^25} {jd:^20.5f} {mjd:^20.5f}")

# Convertir de string a Time y a JD
print(f"\n" + "="*60)
print("Conversión: String -> Time -> JD -> String")
print("="*60)

# String -> Time
fecha_especifica = "2029-06-15 14:30:45"
t_especifica = Time(fecha_especifica, scale="utc")

# Time -> JD
jd_especifico = t_especifica.jd

# JD -> Time -> String
t_desde_jd = Time(jd_especifico, format="jd", scale="utc")
fecha_reconvertida = t_desde_jd.iso

print(f"Original:      {fecha_especifica}")
print(f"JD:            {jd_especifico:.10f}")
print(f"Reconvertido:  {fecha_reconvertida}")

# Diferencias en el análisis
print(f"\n" + "="*60)
print("Diferencias entre tiempos en unidades útiles")
print("="*60)

t_inicio = Time("2029-04-01 00:00:00", scale="utc")
t_fin = Time("2029-05-11 15:30:00", scale="utc")

delta_t = t_fin - t_inicio

print(f"Inicio: {t_inicio.iso}")
print(f"Fin:    {t_fin.iso}")
print(f"\nDiferencia en diferentes unidades:")
print(f"  Días:      {delta_t.jd:.6f}")
print(f"  Horas:     {delta_t.jd * 24:.6f}")
print(f"  Minutos:   {delta_t.jd * 24 * 60:.6f}")
print(f"  Segundos:  {delta_t.sec:.6f}")
```

```python
# ================================================================
# 8. COMPARATIVA: datetime vs astropy.time
# ================================================================
print("\n" + "="*60)
print("8. COMPARATIVA: datetime vs astropy.time")
print("="*60)

fecha = "2029-04-13 12:30:00"

# Con datetime
print(f"\n### DATETIME ###")
dt = datetime.fromisoformat(fecha)
print(f"Crear:     datetime.fromisoformat('{fecha}')")
print(f"Valor:     {dt}")
print(f"ISO:       {dt.isoformat()}")
print(f"Suma:      {dt + timedelta(days=1)}")
print(f"Resta:     {(datetime(2029, 4, 15, 12, 30, 0) - dt).total_seconds()} segundos")
print(f"Julian:    ❌ NO disponible nativamente")

# Con astropy.time
print(f"\n### ASTROPY.TIME ###")
t = Time(fecha, scale="utc")
print(f"Crear:     Time('{fecha}', scale='utc')")
print(f"Valor:     {t.iso}")
print(f"ISO:       {t.isot}")
print(f"Suma:      {(t + 1*u.day).iso}")
print(f"Resta:     {(Time('2029-04-15 12:30:00', scale='utc') - t).jd} días")
print(f"Julian:    JD = {t.jd:.5f}, MJD = {t.mjd:.5f}")

print(f"\n✓ Para astronomía y cálculos de órbitas, usa ASTROPY.TIME")
print(f"✓ Para aplicaciones generales, datetime es más simple")
```

```python
# ================================================================
# 9. TRUCOS Y CASOS DE USO ÚTILES
# ================================================================
print("\n" + "="*60)
print("9. TRUCOS Y CASOS DE USO ÚTILES")
print("="*60)

# 1. Secuencia de tiempos
print("\n### Crear secuencia de tiempos con intervalo ###")
t_inicio = Time("2029-04-13 00:00:00", scale="utc")
intervalo = 6 * u.hour
secuencia = t_inicio + [i for i in range(5)] * intervalo
print("Cada 6 horas durante 5 periodos:")
for i, t in enumerate(secuencia):
    print(f"  t[{i}] = {t.iso}")

# 2. Ahora mismo (UTC actual)
print("\n### Tiempo actual en UTC ###")
ahora = Time.now()
print(f"Ahora: {ahora.iso}")
print(f"JD:    {ahora.jd}")

# 3. Conversión de formato complejo
print("\n### Formato personalizado de salida ###")
t = Time("2029-04-13 12:30:45", scale="utc")
print(f"ISO:     {t.iso}")
print(f"ISOT:    {t.isot}")
print(f"yday:    {t.yday}")
print(f"fits:    {t.fits}")
print(f"mjd:     {t.mjd:.10f}")

# 4. Diferentes escalas de tiempo
print("\n### Escalas de tiempo diferentes ###")
t_utc = Time("2029-04-13 12:00:00", scale="utc")
print(f"UTC:  {t_utc.iso}")
print(f"TAI:  {t_utc.tai.iso if hasattr(t_utc.tai, 'iso') else t_utc.tai}")

# 5. Estadísticas de múltiples tiempos
print("\n### Operaciones con arrays de tiempos ###")
tiempos_array = Time([
    "2029-04-13 00:00:00",
    "2029-04-13 06:00:00",
    "2029-04-13 12:00:00",
    "2029-04-13 18:00:00",
], scale="utc")

print(f"Array de tiempos: {len(tiempos_array)} elementos")
print(f"JD mín: {tiempos_array.jd.min():.5f}")
print(f"JD máx: {tiempos_array.jd.max():.5f}")
print(f"Duración total: {(tiempos_array.jd.max() - tiempos_array.jd.min()) * 24:.1f} horas")

# 6. Filtrado basado en hora
print(f"\nTiempos después del mediodía:")
for t in tiempos_array:
    iso_str = t.iso  # "2029-04-13 12:00:00.000"
    hora = int(iso_str.split()[1].split(':')[0])  # Tomar la hora
    if hora >= 12:
        print(f"  {t.iso}")
```
